```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns

df = pd.read_csv('nekretnine_cleaned.csv')

# 1. Izračunaj cenu po kvadratu
df["Price per sqm"] = df["Price in EUR"] / df["Size"]

# 2. Prosečna cena po kvadratu za svaku lokaciju
location_avg_price = df.groupby("Location")["Price per sqm"].mean().reset_index()

# 3. Elbow metoda – određivanje optimalnog broja klastera
inertia = []
K_range = range(1, 11)  # Testiramo od 1 do 10 klastera

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(location_avg_price[["Price per sqm"]])
    inertia.append(kmeans.inertia_)

# 4. Prikaz Elbow metode
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, marker='o', linestyle='-')
plt.xlabel("Broj klastera")
plt.ylabel("Inertia")
plt.title("Elbow Metoda za optimalan broj klastera")
plt.show()
```

```python
# 5. Postavi optimalan broj klastera (iz Elbow metode)
optimal_k = 5  # PROMENI OVO NA PRAVI BROJ KLASTERA KOJI VIDIŠ U GRAFU

# 6. Klasterovanje
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
location_avg_price["Cluster"] = kmeans.fit_predict(location_avg_price[["Price per sqm"]])

# 7. Dodaj klastere u originalni DataFrame, ali zadrži Location kolonu
df = df.merge(location_avg_price[["Location", "Cluster"]], on="Location", how="left")
```

```python
# 8. Vizualizacija klastera – Lokacija na X-osi, Cena po kvadratu na Y-osi
plt.figure(figsize=(14, 6))
sns.scatterplot(
    x=location_avg_price["Location"],
    y=location_avg_price["Price per sqm"],
    hue=location_avg_price["Cluster"],
    palette="tab10",
    s=100
)

plt.xticks(rotation=90)  # Rotacija lokacija radi bolje čitljivosti
plt.xlabel("Lokacija")
plt.ylabel("Cena po kvadratu (EUR/m²)")
plt.title("Klasterovanje lokacija na osnovu cene po kvadratu")
plt.legend(title="Klaster")
plt.show()

df.head()
```
